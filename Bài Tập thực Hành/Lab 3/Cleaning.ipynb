# Cleaning

```python
def standardize_text_columns(df):
    df = df.copy()
    for col in df.select_dtypes(include=["object", "string"]).columns:
        df[col] = df[col].astype("string").str.strip()
        df[col] = df[col].replace({
            "": pd.NA,
            "nan": pd.NA,
            "NaN": pd.NA,
            "None": pd.NA
        })
    return df


def fill_with_mode(df, cols):
    df = df.copy()
    for col in cols:
        if col in df.columns and df[col].isna().any():
            mode_value = df[col].mode(dropna=True)
            if len(mode_value) > 0:
                df[col] = df[col].fillna(mode_value.iloc[0])
    return df


def drop_high_missing(df, threshold=0.95, exclude=None):
    df = df.copy()
    exclude = set(exclude or [])
    missing_ratio = df.isna().mean()
    drop_cols = [
        c for c, r in missing_ratio.items()
        if r >= threshold and c not in exclude and c != "label"
    ]
    df = df.drop(columns=drop_cols, errors="ignore")
    return df, drop_cols, missing_ratio


def drop_constant_cols(df):
    df = df.copy()
    constant_cols = [
        c for c in df.columns
        if c != "label" and df[c].nunique(dropna=False) <= 1
    ]
    df = df.drop(columns=constant_cols, errors="ignore")
    return df, constant_cols
```

```python
df_4_6_tidy, dropped_missing_4_6, missing_ratio_4_6 = drop_high_missing(
    df_4_6_tidy,
    threshold=0.95,
    exclude=["OTHER AREA SHIP DIV", "Ship Mode", "SHIP DECISION NO", "SUPPLIER_DIV"]
)

print("Dropped because of high missing ratio (4_6):")
print(dropped_missing_4_6)
```

# Type conversion

```python
def convert_datetime_cols(df, cols):
    df = df.copy()
    for col in cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")
    return df


def convert_string_cols(df, cols):
    df = df.copy()
    for col in cols:
        if col in df.columns:
            df[col] = df[col].astype("string")
    return df
```

## Chia nhóm date/time, num, cate

```python
def split_column_types(df):
    date_cols = df.select_dtypes(include=["datetime64[ns]", "datetimetz"]).columns.tolist()
    num_cols = df.select_dtypes(include=np.number).columns.tolist()
    cat_cols = df.select_dtypes(include=["object", "string", "category"]).columns.tolist()
    
    return date_cols, num_cols, cat_cols
```
