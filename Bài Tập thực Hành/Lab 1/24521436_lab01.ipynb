### k. Vẽ biểu đồ về số ca lây nhiễm nhiều nhất của 3 quốc gia: Vietnam, Indonesia và Philippine trong 2 tháng gồm 01 và tháng 02 năm 2021. 

```python
countries = ["Vietnam", "Indonesia", "Philippines"]
coronaJanVIP = df[(df["ObservationDate"] >= "2021-01-01") & (df["ObservationDate"] <= "2021-01-31") & (df["Country/Region"].isin(countries))]
coronaFebVIP = df[(df["ObservationDate"] >= "2021-01-01") & (df["ObservationDate"] <= "2021-02-28") & (df["Country/Region"].isin(countries))]
def Confirmed_daily(df):
    df_daily = df.groupby(["Country/Region","ObservationDate"])["Confirmed"].sum().reset_index()
    df_daily["NewConfirmedDaily"] = df_daily.groupby("Country/Region")["Confirmed"].diff().fillna(0)
    return df_daily.groupby("Country/Region")["NewConfirmedDaily"].max()
maxJanVIP = Confirmed_daily(coronaJanVIP)
maxFebVIP = Confirmed_daily(coronaFebVIP)

plt.figure(figsize = (20,12))
x = np.arange(len(countries))
width = 0.35
plt.bar(x - width/2, maxJanVIP, width, label="Tháng 1/2021")
plt.bar(x + width/2, maxFebVIP, width, label="Tháng 2/2021")
plt.xticks(x, countries)
plt.legend()
plt.show()
```

- Đầu tiên ta chọn dữ liệu theo thời gian và quốc gia
- Sau đó lập hàm tính toán ra số ca mới hàng ngày rồi tìm ra số ca hàng ngày lớn nhất mới trong tháng 1 và tháng 2
- Cuối cùng là vẽ biểu đồ cột ghép với mỗi nhóm country
- Nhận xét biểu đồ, trong 2 tháng thì giá trị số ca mới cao nhất việt nam và Philipines trong 2 tháng không có thay đổi trong khi indonesia có biến động nhẹ, rõ ràng số ca mới của Việt Nam là rất cao trong khi philip rất thấp và indonesia là khá thấp
